```python
from astropy.io import fits
from astropy.table import Table
import pandas as pd

def fits_to_dataframe(fits_path, ext=1):
    """
    Convert a FITS table extension to a pandas DataFrame.
    
    ext=1 is usually correct for light curve files.
    """
    with fits.open(fits_path) as hdul:
        data = Table(hdul[ext].data)
    
    df = data.to_pandas()
    return df
```
