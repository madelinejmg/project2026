````python
# grab the sectors string from the first matching row
sectors_str = df.loc[df["TIC ID"] == tic_id, "Sectors"].iloc[0]

# parse it into ints
sectors = [int(float(s.strip())) for s in str(sectors_str).split(",") if s.strip()]
print(f'TIC ID {tic_id}: Sector(s) {sectors}')
````
